```python
### import necessary packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.offsetbox import AnchoredText  # for adding some text
```

```python
### load the data using pd directly
link = "/home/jovyan/work/Desktop/CASA0007/Group_Presentation/Code_Visual_part/python_code_visual/AoLR_Data_collection 2017_edit.csv"
df = pd.read_csv(link)
```

```python
### select the required topic column
# adding a column called fires which sum 1 pump fires and 2+ pump fires
df['fires'] = df['1 pump fires'] + df['2+ pump fires']
# use a name_list to select the required column
name_list = ['Ward Code', 'Ward name', 'Density', '65+ age', 'HR buildings', 'Student', 'Over crowding', 'Smokers', 'fires']
df = df[name_list]
```

```python
### correlation scatter plot
# select the numeric column to make the correlation matrix
df_numeric = df.loc[:, 'Density':'fires']
# make the correlation matrix
df_cormatrix = df_numeric.corr()
# add a list to record the correlation coefficient values
corcoef_list = df_cormatrix['fires'].tolist()
```

```python
### scatter plots of the concerns vs fires
fig, axes = plt.subplots(2, 3, figsize = (25, 15))
fig.suptitle('Scatter plot of concerns and total fires', size = 24)


sns.regplot(x = 'Density', y = 'fires', data = df, color = 'blue', marker='x', ax = axes[0, 0])
axes[0, 0].set_title('Density', size = 20)
axes[0, 0].set_xlabel('Density', size = 15)
axes[0, 0].set_ylabel('Total fires', size = 15)
axes[0, 0].grid(linestyle=':', linewidth=0.8)
axes[0, 0].add_artist(AnchoredText(f"correlation = {corcoef_list[0]:.3f}", loc=1,
                   prop=dict(size='18', color='k', weight='bold')))


sns.regplot(x = '65+ age', y = 'fires', data = df, color = 'blue', marker='x', ax = axes[0, 1])
axes[0, 1].set_title('65+ age', size = 20)
axes[0, 1].set_xlabel('65+ age', size = 15)
axes[0, 1].set_ylabel('Total fires', size = 15)
axes[0, 1].grid(linestyle=':', linewidth=0.8)
axes[0, 1].add_artist(AnchoredText(f"correlation = {corcoef_list[1]:.3f}", loc=1,
                   prop=dict(size='18', color='k', weight='bold')))


sns.regplot(x = 'HR buildings', y = 'fires', data = df, color = 'blue', marker='x', ax = axes[0, 2])
axes[0, 2].set_title('HR buildings', size = 20)
axes[0, 2].set_xlabel('HR buildings', size = 15)
axes[0, 2].set_ylabel('Total fires', size = 15)
axes[0, 2].grid(linestyle=':', linewidth=0.8)
axes[0, 2].add_artist(AnchoredText(f"correlation = {corcoef_list[2]:.3f}", loc=1,
                   prop=dict(size='18', color='k', weight='bold')))


sns.regplot(x = 'Student', y = 'fires', data = df, color = 'blue', marker='x', ax = axes[1, 0])
axes[1, 0].set_title('Student', size = 20)
axes[1, 0].set_xlabel('Student', size = 15)
axes[1, 0].set_ylabel('Total fires', size = 15)
axes[1, 0].grid(linestyle=':', linewidth=0.8)
axes[1, 0].add_artist(AnchoredText(f"correlation = {corcoef_list[3]:.3f}", loc=1,
                   prop=dict(size='18', color='k', weight='bold')))


sns.regplot(x = 'Over crowding', y = 'fires', data = df, color = 'blue', marker='x', ax = axes[1, 1])
axes[1, 1].set_title('Over crowding', size = 20)
axes[1, 1].set_xlabel('Over crowding', size = 15)
axes[1, 1].set_ylabel('Total fires', size = 15)
axes[1, 1].grid(linestyle=':', linewidth=0.8)
axes[1, 1].add_artist(AnchoredText(f"correlation = {corcoef_list[4]:.3f}", loc=1,
                   prop=dict(size='18', color='k', weight='bold')))


sns.regplot(x = 'Smokers', y = 'fires', data = df, color = 'blue', marker='x', ax = axes[1, 2])
axes[1, 2].set_title('Smokers', size = 20)
axes[1, 2].set_xlabel('Smokers', size = 15)
axes[1, 2].set_ylabel('Total fires', size = 15)
axes[1, 2].grid(linestyle=':', linewidth=0.8)
axes[1, 2].add_artist(AnchoredText(f"correlation = {corcoef_list[5]:.3f}", loc=1,
                   prop=dict(size='18', color='k', weight='bold')))


plt.show()
```
